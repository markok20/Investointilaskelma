# Selvitettävät asiat

Selvitä näistä investoinneista:

Kummankin investoinnin nykyarvot korkokannalla 10 %.

Kuinka investoinnit käyttäytyvät korkokannoilla 0—20 %. Anna vastaus numeerisessa ja graafisessa muodossa.

Kumpaa investointia suosittelet? Perustele vastauksesi useammasta näkökulmasta.

```python
import numpy as np
import matplotlib.pyplot as plt

# Investointi A

hankintahinta_A = 1000000
tuotot_A = [405000, 405000, 405000]  # Kolme vuotta 405 000 euroa
```

```python
# Investointi B

hankintahinta_B = 600000
tuotot_B = [140000, 140000, 140000, 140000, 140000, 140000]  # Kuusi vuotta 140 000 euroa
```

```python
# Nykyarvon laskentafunktio

def laske_nykya_arvo(hankintahinta, tuotot, korkokanta):
    npv = -hankintahinta  # Hankintahinta on aluksi negatiivinen
    for t, tuotot_vuosi in enumerate(tuotot, start=1):
        npv += tuotot_vuosi / (1 + korkokanta) ** t  # Diskontataan kassavirta
    return npv
```

```python
# Määritellään korkokannat välillä 0-20%
korkokannat = np.linspace(0, 0.2, 21)  
```

```python
# Lasketaan nykyarvot  investoinnille
npv_A = [laske_nykya_arvo(hankintahinta_A, tuotot_A, r) for r in korkokannat]
npv_B = [laske_nykya_arvo(hankintahinta_B, tuotot_B, r) for r in korkokannat]
```

```python
# Tulostetaan nykyarvot 10 %:n korkokannalla
npv_A_10 = laske_nykya_arvo(hankintahinta_A, tuotot_A, 0.1)
npv_B_10 = laske_nykya_arvo(hankintahinta_B, tuotot_B, 0.1)
```

```python
print(f"Investointi A:n nykyarvo 10 %:n korkokannalla: {npv_A_10:.2f} euroa")
print(f"Investointi B:n nykyarvo 10 %:n korkokannalla: {npv_B_10:.2f} euroa")
```

```python
# Piirretään graafi
plt.figure(figsize=(10, 6))
plt.plot(korkokannat * 100, npv_A, label='Investointi A', color='blue')  # Sininen käyrä investoinnille A
plt.plot(korkokannat * 100, npv_B, label='Investointi B', color='green')  # Vihreä käyrä investoinnille B
plt.xlabel('Korkokanta (%)')
plt.ylabel('Nykyarvo (eur)')
plt.title('Nykyarvojen vertailu eri korkokannoilla')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Perustelu suositukselle
if npv_A_10 > npv_B_10:
    print("Suosittelen investointia A, nykyarvo on korkeampi 10 %:n korkokannalla.")
else:
    print("Suosittelen investointia B, nykyarvo on korkeampi 10 %:n korkokannalla.")
```

# Johtopäätökset

Nykyarvon laskemisessa käytetään diskonttokorkoa, joka vie tulevat kassavirrat nykyhetkeen.  

Graafissa sininen käyrä edustaa investointia A ja vihreä käyrä investointia B.

Suosittelen sitä investointia, jonka nykyarvo on korkeampi 10 %:n korkokannalla. 

Tämä korkokanta on yleisesti käytetty esimerkki markkinoiden keskimääräisestä tuotto-odotuksesta. 

Jos investointi A:n nykyarvo on korkeampi 10 %:n korkokannalla, se on taloudellisesti kannattavampi kuin investointi B, ja siksi se on parempi.
