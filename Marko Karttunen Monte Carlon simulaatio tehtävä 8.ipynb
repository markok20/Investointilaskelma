# Monte Carlon simulaatio 24.3.2025 tehtävä 8

Tehtävä:

mainosten määrä l: normaalijakautunut, keskiarvo 1 500, keskihajonta 165

ostotapahtumaan johtavien osuus r: tasaisesti jakautunut, min 1 %, max 5 %
    
voitto tuotetta kohden p: tasaisesti jakautunut, min 47 €, max 53 €
    
kustannus mainosta kohden c: tasaisesti jakautunut, min 0,20 €, max 0,80 €
    
yrityksen kiinteät kustannukset h: 800 euroa kuukaudessa (kiinteä).

Käytä Monte Carlo -simulointia ja arvioi sen avulla yrityksen tulos kuukausitasolla. 


```python
# kirjastot sisään

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Määritellään simulaatioiden määrä
n_simulations = 10000
```

```python
# määritellään muuttujien jakauamat

l = np.random.normal(1500, 165, n_simulations)  # Mainosten määrä 
r = np.random.uniform(0.01, 0.05, n_simulations)  # Ostotapahtuman todennäköisyys
p = np.random.uniform(47, 53, n_simulations)  # Voitto per tuote 
c = np.random.uniform(0.20, 0.80, n_simulations)  # Mainoksen kustannus 
h = 800  # Kiinteät kustannukset 
```

```python
# Lasketaan tulos Monte Carlo -simulaatiolla
profit = (l * r * p) - (l * c + h)
```

```python
# Tulosten visualisointi
plt.figure(figsize=(10, 5))
plt.hist(profit, bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(profit), color='red', linestyle='dashed', linewidth=2, label=f'Keskiarvo: {np.mean(profit):.2f} €')
plt.xlabel('Kuukausittainen tulos (€)')
plt.ylabel('Frekvenssi')
plt.title('Monte Carlo -simulaation tulosjakauma')
plt.legend()
plt.grid()
plt.show()
```

```python
# Tulostetaan keskeiset tunnusluvut
print(f"Keskiarvoinen kuukausittainen tulos: {np.mean(profit):.2f} €")
print(f"Mediaani: {np.median(profit):.2f} €")
print(f"Alakvartiili (25%): {np.percentile(profit, 25):.2f} €")
print(f"Yläkvartiili (75%): {np.percentile(profit, 75):.2f} €")
print(f"Tappion todennäköisyys: {np.mean(profit < 0) * 100:.2f} %")
```

# Monte Carlo -simulaation sanallinen yhteenve

Arvioidaan yrityksen kuukausittaista tulosta ottaen huomioon epävarmuustekijät, kuten mainosten määrän, ostotapahtumaan johtavien osuus, yksittäisen tuotteen voitto sekä mainosten kustannukset. 

Suoritettiin 10 000 satunnaisotantaa, joissa jokainen muuttuja generoidaan ennalta määrättyjen jakaumien mukaisesti. 

Näiden avulla laskettiin mahdollinen kuukausittainen tulos käyttämällä kaavaa:

Tulos = l * r * p - (l * c + h)

Tulosten jakauma visualisoitiin histogrammilla.

Keskeiset tunnusluvut, kuten keskiarvo, mediaani ja tappion todennäköisyys, esitetään erikseen. 

Näin yritys voi arvioida liiketoimintansa kannattavuutta ja mahdollisia riskejä paremmin.
